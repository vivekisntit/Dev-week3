# Data Cleaning

```python
df["Age"]=[25, 30, 35, np.nan, 29, 25]
df
```

```python
df.isnull().sum()
```

```python
df.dropna(how="any")
```

```python
df.dropna(how="all") #if all the values in a row are null
```

```python
df.fillna(0)
```

```python
df["Age"].fillna(df["Age"].mean())
```

```python
df["Salary"].fillna(df["Salary"].median())
```

```python
df.fillna(method="ffill")  
```

```python
df_dup=df[df.duplicated(keep="last")]
df_dup
```

```python
df=df.drop_duplicates()
df
```
